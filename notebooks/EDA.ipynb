```python
from dotenv import load_dotenv
import pandas as pd
import os
from sqlalchemy import create_engine

load_dotenv()

engine = create_engine(
    f"postgresql://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
)
```

```python
query = ''' SELECT * FROM customers LIMIT 5 '''
df = pd.read_sql(query, engine)
df
```

```python
query = ''' SELECT COUNT(DISTINCT customer_id)
            FROM customers '''
df = pd.read_sql(query, engine)
df
```

```python
query = ''' SELECT COUNT(DISTINCT customer_unique_id)
            FROM customers '''
df = pd.read_sql(query, engine)
df
```

we can say that customers can have multiple 'customer_id' but only one 'customer_unique_id'

```python
query = ''' SELECT COUNT(order_id),
            COUNT(DISTINCT order_id) AS unique_order_ids
            FROM orders '''
df = pd.read_sql(query, engine)
df
```

all order_ids are distinct

```python
query = ''' SELECT DISTINCT order_status
            FROM orders; '''
df = pd.read_sql(query, engine)
df
```

```python
query = ''' SELECT COUNT(seller_id),
            COUNT(DISTINCT seller_id) AS Unique_sellers
            FROM sellers;  '''
df = pd.read_sql(query, engine)
df
```

```python
query = ''' SELECT COUNT(product_id)AS total_products,
            COUNT(DISTINCT product_id) AS unique_products
            FROM products;  '''
df = pd.read_sql(query, engine)
df
```

we can say all products are unique but they may have same product categories.
